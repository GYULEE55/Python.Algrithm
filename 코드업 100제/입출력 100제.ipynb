6095 : [기초-리스트] 바둑판에 흰 돌 놓기(설명)(py)

```python
d = [[0] * 20 for _ in range(20)]  # 19x19 바둑판 (0번 안 씀)

n = int(input())
for _ in range(n):
    x, y = map(int, input().split())
    d[x][y] = 1  # 흰 돌 놓기

for i in range(1, 20):
    for j in range(1, 20):
        print(d[i][j], end=' ')
    print()
```

6096 : [기초-리스트] 바둑알 십자 뒤집기(py)

```python
d = []
for i in range(20):
    d.append([])
    for j in range(20):
        d[i].append(0)

# 바둑판 입력 (19x19)
for i in range(1, 20):
    row = input().split()
    for j in range(1, 20):
        d[i][j] = int(row[j-1])

n = int(input())
for _ in range(n):
    x, y = map(int, input().split())

    # 가로줄 뒤집기
    for j in range(1, 20):
        d[x][j] = 1 - d[x][j]   # 0→1, 1→0

    # 세로줄 뒤집기
    for i in range(1, 20):
        d[i][y] = 1 - d[i][y]

# 출력
for i in range(1, 20):
    for j in range(1, 20):
        print(d[i][j], end=' ')
    print()
```

6098 : [기초-리스트] 성실한 개미(py)

```python
array = []

for i in range(10):
    array.append(list(map(int, input().split())))

x, y = 1, 1

while True:
    if (array[x][y] == 0):
        array[x][y] = 9
    elif (array[x][y] == 2):
        array[x][y] = 9
        break

    if ((array[x][y+1] == 1 and array[x+1][y] == 1)):
        break

    if (array[x][y+1] != 1):
        y = y + 1
    elif (array[x+1][y] != 1):
        x = x + 1

for i in range(10):
    for j in range(10):
        print(array[i][j], end=' ')
    print()
```
